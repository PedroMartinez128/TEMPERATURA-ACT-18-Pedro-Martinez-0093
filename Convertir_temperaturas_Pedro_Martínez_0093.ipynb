## 1. Fórmulas de Conversión

Para realizar estas conversiones, utilizamos relaciones lineales basadas en los puntos de congelación y ebullición del agua.

* **De Celsius a Fahrenheit:** multiplicamos por 1.8 (o $9/5$) y sumamos 32.
    $$°F = (°C \times 1.8) + 32$$

* **De Fahrenheit a Celsius:** restamos 32 y dividimos el resultado entre 1.8.
    $$°C = \frac{°F - 32}{1.8}$$

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo A: Convertir 25°C a Fahrenheit (Temperatura ambiente)
1.  **Identificar datos:** Tenemos $°C = 25$.
2.  **Sustituir en la fórmula:** $°F = (25 \times 1.8) + 32$.
3.  **Multiplicar:** $25 \times 1.8 = 45$.
4.  **Sumar:** $45 + 32 = 77$.
* **Resultado:** **77°F**.

### Ejemplo B: Convertir 98.6°F a Celsius (Temperatura corporal promedio)
1.  **Identificar datos:** Tenemos $°F = 98.6$.
2.  **Sustituir en la fórmula:** $°C = (98.6 - 32) / 1.8$.
3.  **Restar:** $98.6 - 32 = 66.6$.
4.  **Dividir:** $66.6 / 1.8 = 37$.
* **Resultado:** **37°C**.

## 3. Código en Python

Este script es interactivo y sigue exactamente la estructura de funciones que solicitaste:

```python
def convertirCF():
    """Convierte grados Celsius a Fahrenheit"""
    celsius = float(input("\nIngrese los grados Celsius: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"--- {celsius}°C equivale a {fahrenheit:.2f}°F ---")

def convertirFC():
    """Convierte grados Fahrenheit a Celsius"""
    fahrenheit = float(input("\nIngrese los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"--- {fahrenheit}°F equivale a {celsius:.2f}°C ---")

def menu():
    """Muestra el menú principal y gestiona la lógica"""
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")

        print("3. Salir")
        
        opcion = input("Seleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

# Punto de entrada del programa

    menu()
```

### Explicación breve del código:
* **`while True`**: Crea un bucle infinito para que el menú se repita hasta que tú decidas salir.
* **`float()`**: Permite que el usuario ingrese números con decimales (importante en ciencia).
* **`f-strings`**: Usamos `f"{valor:.2f}"` para redondear el resultado a 2 decimales y que se vea más limpio.
* **`break`**: Rompe el ciclo inmediatamente cuando el usuario elige la opción 3.

````python
def convertirCF():
    """Convierte grados Celsius a Fahrenheit"""
    celsius = float(input("\nIngrese los grados Celsius: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"--- {celsius}°C equivale a {fahrenheit:.2f}°F ---")

def convertirFC():
    """Convierte grados Fahrenheit a Celsius"""
    fahrenheit = float(input("\nIngrese los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"--- {fahrenheit}°F equivale a {celsius:.2f}°C ---")

def menu():
    """Muestra el menú principal y gestiona la lógica"""
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a grados Fahrenheit")
        print("2. Convertir grados Fahrenheit a grados Centígrados")
        print("3. Salir")

        opcion = input("Seleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("Programa realizado por Pedro Martinez NL = 34")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

if __name__ == "__main__":
    menu()
````
